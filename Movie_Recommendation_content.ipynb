```python
df.isnull().sum()
```

```python
features = ['keywords', 'cast', 'genres', 'director']

for feature in features:
    df[feature] = df[feature].fillna("")
# df.isnull().sum()
```

```python
def combine(row):
    try:
        return row['keywords'] + " " + row['cast'] + " " + row['genres'] + " " + row['director'] 
    except:
        print("ERROR----->", row)

df['combine_cols'] = df.apply(combine,axis=1)
df.combine_cols.head()
```

```python
vect = CountVectorizer()
```

```python
count_matrix = vect.fit_transform(df['combine_cols'])
```

```python
# print(vect.get_feature_names())
print(count_matrix.toarray())
#here we can see 4803 documents nd 14845 features and we hve only 97547 values in matrix remaining are 0
count_matrix   
```

```python
#give similarity based on 
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim.shape)  #we convert to index as document and col as document as well
cosine_sim
```
