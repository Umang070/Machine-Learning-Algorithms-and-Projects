```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
data = load_breast_cancer()
#data.data.shape --->(569,30)
#data.target.shape -->(569,1)

X = data.data
y = data.target
```

```python
X_train,y_train,X_test,y_test = train_test_split(X,y,random_state=1)
 
#prediction -- > return all models predicted op in dataframe
clf = LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None,predictions=True) 
models,prediction=clf.fit(X_train,y_train,X_test,y_test)
```
